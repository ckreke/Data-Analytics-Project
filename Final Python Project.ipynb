```python
import pandas as pd
from pandas import Series,DataFrame
import numpy as np
%pylab inline
```

```python
df = pd.read_csv('athlete_events.csv', dtype=unicode)
print len(df)
df.head()
```

```python
df.dropna(subset = ['Age','Height','Weight'],inplace=True)
df['Age'] = df['Age'].astype(float)
df['Height'] = df['Height'].astype(float)
df['Weight'] = df['Weight'].astype(float)
df['Year'] = df['Year'].astype(int)
len(df)
```

```python
def periodSplit(y):
    if (y < 1926):
        period = '1early'
    elif (y < 1956):
        period = '2mid_early'
    elif ( y < 1986):
        period = '3early_modern'
    else:
        period = '4modern'
        
    return period

df['Period'] = df['Year'].map(periodSplit)

df.head()
```

```python
country_period = df.groupby(['NOC','Period'])['Age','Height','Weight'].mean()

country_period.head()
```

```python
country_stats = country_period.unstack()
country_stats.dropna()
```
