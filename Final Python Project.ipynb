```python
import pandas as pd
from pandas import Series,DataFrame
import numpy as np
%pylab inline
```

```python
raw_df = pd.read_csv('athlete_events.csv', dtype=unicode)
raw_df['Year'] = raw_df['Year'].astype(int)
raw_df['Age'] = raw_df['Age'].astype(float)
raw_df['Height'] = raw_df['Height'].astype(float)
raw_df['Weight'] = raw_df['Weight'].astype(float)
print len(raw_df)
raw_df.head()
```

```python
df = raw_df.copy()
df.dropna(subset = ['Age','Height','Weight'],inplace=True)
len(df)
```

```python
def periodSplit(y):
    if (y < 1926):
        period = '1early'
    elif (y < 1956):
        period = '2mid_early'
    elif ( y < 1986):
        period = '3early_modern'
    else:
        period = '4modern'
        
    return period

df['Period'] = df['Year'].map(periodSplit)

df.head()
```

```python
country_period = df.groupby(['NOC','Period'])['Age','Height','Weight'].mean()

country_period.head()
```

```python
country_stats = country_period.unstack()
country_stats.dropna()
```

```python
df1 = raw_df[(raw_df['Year'] >= 1950) &(raw_df['Sex'] == 'M')]
df2 = df1[['Name','NOC','Sport','Medal']] # Sport or event??
# do we want an athlete to be counted every olympics he attends, or just one time for whole life??
df2['Unique Athlete'] = df2['Name'] + df2['NOC'] + df2['Sport']
# do we want to consider only sports that started after ~1950? Records from before then might not include all athletes who competed.
df2.head()
```

```python
winnerCount = df2.groupby(['Sport'])['Medal'].count()
winnerCount.head()
```

```python
entrantCount = df2.groupby(['Sport'])['Unique Athlete'].count()
entrantCount.head()
```

```python
df3 = DataFrame(entrantCount.values, index = entrantCount.index.values)
df3.columns = ['Entrants']
df3['Winners'] = winnerCount.values
df3['Percent Winners'] = df3['Winners'] * 100.0 / df3['Entrants']
df3.sort_values(by='Winners', inplace=True)
df3.iloc[:30]
```

# Clustering

```python
from patsy import dmatrices
import warnings
warnings.filterwarnings('ignore')
from sklearn.cluster import KMeans
from sklearn.preprocessing import scale
```

```python
sport_stats = df1.groupby('Sport')[['Age','Height','Weight']].mean()
sport_stats.head()
```

```python
df5['Avg Age'] = sport_stats['Age']
df5['Avg Height'] = sport_stats['Height']
df5['Avg Weight'] = sport_stats['Weight']
df5.head()
```

```python
X = scale(df5[['Avg Height', 'Avg Weight']])

seed(1)
sum_of_distances = {}
for num_clusters in range(1,11):
    kmeans = KMeans(init='k-means++',n_clusters=num_clusters, random_state=1)
    kmeans.fit(X)
    sum_of_distances[num_clusters] = kmeans.inertia_
Series(sum_of_distances).sort_index().plot()
```

```python
kmeans = KMeans(init='k-means++',n_clusters=4, random_state=1)
kmeans.fit(X)
for cluster in range(4):
    mask = (kmeans.labels_ == cluster)
    result = df5[mask]
    print result.index.values, result['Avg Height'].mean(), result['Avg Weight'].mean()
    print 
```
